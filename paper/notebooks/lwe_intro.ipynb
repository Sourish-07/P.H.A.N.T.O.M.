# SVP Demo
Toy example showing lattice generated by a basis.

```

import numpy as np
import matplotlib.pyplot as plt

basis = np.array([[3,1],[2,2]])
pts = []
for x in range(-5,6):
    for y in range(-5,6):
        pts.append(x*basis[0] + y*basis[1])

pts = np.array(pts)
plt.scatter(pts[:,0], pts[:,1])
plt.title("Lattice generated by basis")
plt.show()
```
